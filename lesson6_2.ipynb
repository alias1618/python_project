```python
#台灣樂透
for _ in range(4):
    import random
    lot = set()
    while(len(lot) < 7):
        lot.add(random.randint(1,49))
    lot_list = list(lot)
    specialNum = lot_list.pop()

    for item in lot_list:
        print(item, end=' ')
    print(f'特別號:{specialNum}')
```

```python
setNum = int(input('請輸入大樂透選號組數'))
for _ in range(setNum):

    lot = set()
    while(len(lot) < 7):
        lot.add(random.randint(1,49))
    lot_list = list(lot)
    specialNum = lot_list.pop()

    for item in lot_list:
        print(item, end=' ')
    print(f'特別號:{specialNum}')
```

```python
#自訂的function
def big_lot():
    lot = set()
    while(len(lot) < 7):
        lot.add(random.randint(1,49))
    lot_list = list(lot)
    specialNum = lot_list.pop()

    for item in lot_list:
        print(item, end=' ')
    print(f'特別號:{specialNum}')

setNum = int(input('請輸入大樂透選號組數'))
for _ in range(setNum):
    big_lot()
```
